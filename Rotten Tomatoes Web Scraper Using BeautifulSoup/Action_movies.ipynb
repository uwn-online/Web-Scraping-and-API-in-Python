Web Scraper with Beautiful Soup
* Inspect the page
* Obtain HTML
* Parse HTML (html.parser, lxml, html5lib)
* Intialise BeautifulSoup Object
* Export HTML to file for inspection

```python
# Import libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://editorial.rottentomatoes.com/guide/140-essential-action-movies-to-watch-now/" # Connect to webpage
response = requests.get(url)   # Returns response using the GET method.
response.status_code   # Confirms if connection is successful.
```

```python
# Instantiate HTML

html = response.content
html
```

```python
# Create BeautifulSoup object.

bs = BeautifulSoup(html, 'html.parser')
bs
```

```python
# Export HTML to local file for inspection.
with open('rotten_tomatoes.html', 'wb') as file:
    file.write(bs.prettify('utf-8'))
```

```python
movie_info = bs.find_all('div', {'class': 'col-sm-18 col-full-xs countdown-item-content'}) # Returns all div class containing movie info.
movie_info
```

```python
headings = [div.find("h2") for div in movie_info] # Returns all h2 in the html structure that stores the title, year, scores.
headings
```

```python
# To extract the title.
movie_title = [h.find('a').string for h in headings] # Finds all titles in h2 and converts to string.
movie_title
```

Etract the years
* Find all year in sub class 'start-year'
* Convert them to string
* Strip parentheses
* Convert to integer

```python
# To extract the years
movie_year = [h.find('span', class_ = 'start-year').string for h in headings]
movie_year
```

```python
# Strip parentheses using .strip('()') method.
movie_year =[year.strip('()') for year in movie_year]
movie_year
```

```python
# Convert to integers from string.
movie_year = [int(year) for year in movie_year]
movie_year
```

To Extract the Score
* Find Score in subclass tMeterScore in the heading (h2)
* Strip the %
* Convert to integer

```python
movie_score = [h.find('span', class_='tMeterScore').string for h in headings] # Returns movie scores stored as subclass within the h2.
movie_score
```

```python
movie_score = [score.strip('%') for score in movie_score] # Strips '%' in each score item in movie_score.
movie_score
```

```python
movie_score = [int(score) for score in movie_score]  # Converts to integer.
movie_score
```

Text Preprocessing.
* Find all consensus in class stored in div tags
* Return each consensus text
* Strip white spaces and other irrelevant characters


```python
consensus = [div.find("div", {"class": "info critics-consensus"})for div in movie_info] # Extract the critics-consensus div from each movie_info block.
consensus
```

```python
[texts.text.strip() for texts in consensus] # Extract the text in consensus.
```

From the output above, we have a common phrase 'Critics Consensus' associated with our texts. 

```python
common_phrase = 'Critics Consensus: '
common_len = len(common_phrase)
common_len  # Returns the length of common phrase.
```

```python
consensus_txt = [texts.text[common_len:].strip() for texts in consensus] # Strips the common phrase using index slicing.
consensus_txt
```

```python
# Extract director of each movie
movie_director = [div.find('div', {'class': 'director'}) for div in movie_info]
movie_director

# Since the directors are stored in a ref
movie_director = [director.find('a').string for director in movie_director] # Converts href to string.
movie_director
```

```python
cast = [div.find('div', {'class': 'cast'}) for div in movie_info] # Extract the cast section for each movie block.
cast

crew = cast[0].find_all('a') # Extract all cast member links within each cast section.
crew

cast_names = [cast.string for cast in crew]
cast_names
```

```python
# Extract all cast crew for the 140 action movies

casting = []
for c in cast:
    crew = c.find_all('a') # Locate all cast member links within the cast block.


    cast_names = [member.string for member in crew] # Extract the cast member names from each link.


    result = ', '.join(cast_names) # Convert the list of cast names into a single comma-spearated string.


    casting.append(result) # Store the formatted cast list for this movie.

casting
```

Export to Pandas DataFrame

```python
essential_action_movies = pd.DataFrame()
pd.set_option('display.max_colwidth', None) # Removes truncation.

essential_action_movies['Title'] = movie_title
essential_action_movies['Release Year'] = movie_year
essential_action_movies['Score'] = movie_score
essential_action_movies['Director(s)'] = movie_director
essential_action_movies['Critics Consensus'] = consensus_txt
essential_action_movies['Casting'] = casting

essential_action_movies
```

```python
# Export to csv file
essential_action_movies.to_csv('essential_action_movies.csv', index = False, header = True)
```
