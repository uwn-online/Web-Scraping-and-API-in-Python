The documentation for itunes can be found here:
* https://affiliate.itunes.apple.com/resources/documentation/itunes-store-web-service-search-api/

```python
# Import the relevant modules.
import requests
import json
```

```python
url = "https://itunes.apple.com/search"  # Base url.
parameters =  {"term": "the beatles", "country": "us", "limit":200} # Specifies lookup values.
```

```python
# Initiate a request to the website
response = requests.get(url, params=parameters) # Initiates a get request with the website.
response.status_code
```

```python
info = response.json() # Returns the content of the website.
print(json.dumps(info, indent=4))
```

```python
print(json.dumps(info['results'][0], indent=4)) # Details of first output.
```

To print the data of interest using list comprehension.


```python
for result in info['results']:  # This returns the track names and ignores where there are no trackname.
    track = result.get("trackName")
    if track:
        print(track)
```

To export to DataFrame

```python
import pandas as pd
```

```python
df = pd.DataFrame(info['results'])
df.head()
```

```python
df.columns # Inspect the columns.
```

```python
# To filter the relevant columns. 
columns_needed =[
    "trackName",
    "country",
    "collectionPrice",
    "trackPrice",
    "trackRentalPrice",
    "collectionHdPrice",
    "trackHdPrice",
    "trackHdRentalPrice",
    "releaseDate",
    "trackExplicitness",
    "genres",
    "contentAdvisoryRating",
    "trackCount",
    "isStreamable"
]
```

```python
df = df[columns_needed] # Returns DataFrame for relevant columns.
```

```python
df.tail()
```

Exploratory Data Analysis & Data Cleaning

```python
df.info() # Information (COLUMN NAMES, NUMBER OF NULL VALUES, DATA TYPES) about the dataset.
```

```python
# To convert release date to a proper datetime column.
df['releaseDate'] = pd.to_datetime(df["releaseDate"]).dt.date
```

```python
df["releaseDate"]
```

```python
df["trackCount"] = df["trackCount"].astype("Int64") # Converts track count to integers.
```

```python
df
```

```python
df.info()
```

```python
# Given the percentage of missing values in columns, trackRentalPrice, trackHdPrice, and trackHdRentalPrice will be dropped.
df = df.drop(["trackRentalPrice", "trackHdPrice", "trackHdRentalPrice"], axis=1)
```

```python
df
```

```python
df.describe(include='object') # Gives a categorical summary of the dataset.
```

```python
# isStreamable is expected to contain either "True" or "False" values.
df["isStreamable"] = df["isStreamable"].replace({True: True, False: False}).fillna(False)
```
